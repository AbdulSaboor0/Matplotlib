```python
import matplotlib.pyplot as plt
```

```python
Date = [1,3,5,7,9]
Cost = [100,120,130,150,112,]
```

### **Plot Function**

---



```python
plt.title('Product Cost')
plt.xlabel('Date')
plt.ylabel('Cost')
plt.plot (Date, Cost, color = 'Green', linewidth = 2, linestyle = 'dashed')
plt.show()
```

```python
### Marker Function

### If you just want to plot the points put (linestyle = '')

### Note : We can also specify a hexadecimal number for color "#004400" (RGB Color)


plt.plot(Date, Cost, marker = 'o', markersize = 7, color = 'red', linestyle = ''
         )
plt.show()
```

```python
### We can also perform the above code in a short form without using the 'marker' function
### We can also change the color of the market by a single letter here 'y'/'g'/'r'etc
plt.plot(Date, Cost,'Dy')
```

```python
### We can control the transparency of the graph by the alpha parameter(range = 0-1)


plt.plot(Date, Cost,'+g--', alpha = 0.5)
```

### **Labels, Legends and Grids**

---



```python

month = [1,2,3,4,5,6,7]
temp_Shibuya = [20,21,23,22,25,26,27]
temp_Kashmir = [15,17,19,18,20,21,18]
temp_Delhi = [30,31,32,33,35,39,40]


plt.title('City Temperatures')
plt.xlabel('Months')
plt.ylabel('Temperature(C)')

### Plotting Multiple Graphs

plt.plot(month, temp_Shibuya, marker = '+', label = 'Shibuya')
plt.plot(month, temp_Kashmir, marker = 'o',label= 'Kashmir')
plt.plot(month, temp_Delhi, marker = "d",label = 'Delhi')

plt.legend(shadow = True, loc = 'best', fontsize = 'large')

plt.grid(color = 'm', alpha = 0.2)
```

### **Plotting Multiple Graphs**

---



```python
 a= [12,10,8,5,3]
 b = [0,3,6,12,30]
 c = [1,2,3,4,5]


### Creating the dimensions for multiple subplots

 plt.figure(figsize = (10,5))

 ### Defining the position of our first plot ( "subplot(no. of rows, no of columns, Position )")

 plt.subplot(1,2,1)
 plt.plot(c,a, marker= "o", color ='green')
 plt.title('First Plot')
 plt.xlabel('x axis')
 plt.ylabel('Y axis')
plt.grid(True, color ="yellow", alpha = 0.3)


 plt.subplot(1,2,2)
 plt.plot(c,b, marker = 'd', color = 'red')
 plt.title('Second plot')
 plt.xlabel('x axis')
 plt.ylabel('y axis')
 plt.grid(True, color = 'yellow', alpha = 0.3)

 plt.show()
```

### **Bar Graphs** ðŸ˜‹

---



```python
company  = ['Amazon', "Walmart", 'Blackrock', 'Brainwave']
revenue = [3,2,8,4]

```

```python
### Normal BarGraph

plt.figure(figsize = (4,4))
plt.title('Company wise yearly revenue')
plt.xlabel('Company')
plt.ylabel('Revenue($B)')
plt.bar(company, revenue, label = 'Revenue', width = 0.5)
plt.legend()

plt.show()
```

```python
import numpy as np


### Multiple bargraphs side by side

category = ['A','B','C','D','E']
value = [3,5,7,4,9]
value2 = [2,4,5,2,5]


# label locations, evenly spread to serve as x axis, we used np becoz normal list of values will give an error in the plot
x = np.arange(len(category))


plt.figure(figsize= (3,3))

### shifting the bars right and left by half of the width
plt.bar(x-0.2, value, width = 0.4, label = 'Value')
plt.bar(x+0.2, value2, width = 0.4, label = 'Value 2')


### Changing the x axis labels to cetegories
plt.xticks(x, category)
plt.xlabel('Category')
plt.ylabel('Value')

plt.legend()
plt.show()
```
