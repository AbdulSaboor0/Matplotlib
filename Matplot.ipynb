```python
import matplotlib.pyplot as plt
```

```python
Date = [1,3,5,7,9]
Cost = [100,120,130,150,112,]
```

### **Plot Function**

---



```python
plt.title('Product Cost')
plt.xlabel('Date')
plt.ylabel('Cost')
plt.plot (Date, Cost, color = 'Green', linewidth = 2, linestyle = 'dashed')
plt.show()
```

```python
### Marker Function

### If you just want to plot the points put (linestyle = '')

### Note : We can also specify a hexadecimal number for color "#004400" (RGB Color)


plt.plot(Date, Cost, marker = 'o', markersize = 7, color = 'red', linestyle = ''
         )
plt.show()
```

```python
### We can also perform the above code in a short form without using the 'marker' function
### We can also change the color of the market by a single letter here 'y'/'g'/'r'etc
plt.plot(Date, Cost,'Dy')
```

```python
### We can control the transparency of the graph by the alpha parameter(range = 0-1)


plt.plot(Date, Cost,'+g--', alpha = 0.5)
```

### **Labels, Legends and Grids**

---



```python

month = [1,2,3,4,5,6,7]
temp_Shibuya = [20,21,23,22,25,26,27]
temp_Kashmir = [15,17,19,18,20,21,18]
temp_Delhi = [30,31,32,33,35,39,40]


plt.title('City Temperatures')
plt.xlabel('Months')
plt.ylabel('Temperature(C)')

### Plotting Multiple Graphs

plt.plot(month, temp_Shibuya, marker = '+', label = 'Shibuya')
plt.plot(month, temp_Kashmir, marker = 'o',label= 'Kashmir')
plt.plot(month, temp_Delhi, marker = "d",label = 'Delhi')

plt.legend(shadow = True, loc = 'best', fontsize = 'large')

plt.grid(color = 'm', alpha = 0.2)
```
